## Additional Features

### Daily Returns

```python
dataframes = [dataframe_eth]

# Concatenate the dataframes along the columns (axis=1)
merged_df = pd.concat(dataframes, axis=1)

# Calculate the daily percentage returns for each cryptocurrency
returns_df = merged_df.pct_change().dropna()

# Plot the daily percentage returns as line charts
plt.figure(figsize=(12, 6))
for column in returns_df.columns:
    plt.plot(returns_df.index, returns_df[column], label=column)
plt.xlabel('Date')
plt.ylabel('Daily Percentage Return')
plt.title('Daily Percentage Returns of Cryptocurrencies')
plt.legend()
plt.grid(True)
plt.show()

# Plot the daily percentage returns as histograms
plt.figure(figsize=(12, 6))
returns_df.hist(bins=30, alpha=0.6)
plt.xlabel('Daily Percentage Return')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Percentage Returns of Cryptocurrencies')
plt.grid(True)
plt.show()
```

### Moving Averages

```python
dataframes = [dataframe_eth]

# Concatenate the dataframes along the columns (axis=1)
merged_df = pd.concat(dataframes, axis=1)

# Plot the scatter plot or bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['market_cap'], merged_df['total_volume'], s=50, alpha=0.5)
plt.xlabel('Market Cap (SGD)')
plt.ylabel('Trading Volume (SGD)')
plt.title('Market Cap vs. Trading Volume')
plt.grid(True)
plt.show()
```

### Correlation Heatmap

```python
dataframes = [dataframe_eth]

# Concatenate the dataframes along the columns (axis=1)
merged_df = pd.concat(dataframes, axis=1)

# Compute the correlation matrix
correlation_matrix = merged_df.corr()

# Plot the correlation heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
